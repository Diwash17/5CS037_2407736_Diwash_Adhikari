Task - 3 - Outlier Detection:

```python
score_q1 = south_asian_data['score'].quantile(0.25)
score_q3 = south_asian_data['score'].quantile(0.75)
score_iqr = score_q3 - score_q1
```

```python
score_lower_bound = score_q1 - 1.5 * score_iqr
score_lower_bound
```

```python
score_upper_bound = score_q3 + 1.5 * score_iqr
score_upper_bound
```

```python
gdp_q1 = south_asian_data['log_gdp_per_capita'].quantile(0.25)
gdp_q3 = south_asian_data['log_gdp_per_capita'].quantile(0.75)
gdp_iqr = gdp_q3 - gdp_q1
```

```python
gdp_lower_bound = gdp_q1 - 1.5 * gdp_iqr
gdp_lower_bound
```

```python
gdp_upper_bound = gdp_q3 + 1.5 * gdp_iqr
gdp_upper_bound
```

```python
outliers = south_asian_data[
    (south_asian_data["score"] < score_lower_bound)
    | (south_asian_data["score"] > score_upper_bound) |
    (south_asian_data["log_gdp_per_capita"] < gdp_lower_bound)
     | (south_asian_data["log_gdp_per_capita"] > gdp_upper_bound)
]
```

```python
outliers[['country_name','score','log_gdp_per_capita']]
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(south_asian_data["log_gdp_per_capita"], south_asian_data["score"], color="lightgreen", label="Non-Outliers")
plt.scatter(outliers["log_gdp_per_capita"], outliers["score"], color="red", label="Outliers")
plt.title("Outlier Detection: GDP per Capita vs Happiness Score (South Asia Region )", fontsize=15)
plt.xlabel("Log GDP per Capita", fontsize=12)
plt.ylabel("Happiness Score", fontsize=12)
plt.legend()
plt.grid()
plt.tight_layout()
plt.savefig('Outlier_detection_scatterplot1.png', bbox_inches='tight',dpi=80)
plt.show()
```

Afghanistan is a potential outlier in the South Asian region.

Characteristics of Outliers

*Extreme Values*: Countries significantly deviating from the regional average in both GDP and scores stand out as clear outliers.

Impact on Regional Averages:

*Misleading Trends*: Outliers can overshadow key drivers of happiness in typical countries.
